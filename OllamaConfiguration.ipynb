put ollama in background

```python
import subprocess
import time

print("Starting Ollama server in the background")
process = subprocess.Popen(
    ["ollama", "serve"],
    stdout=subprocess.DEVNULL, # Hides the server logs
    stderr=subprocess.DEVNULL  # Hides server errors
)

time.sleep(3)
print("Server should be running.")
```
